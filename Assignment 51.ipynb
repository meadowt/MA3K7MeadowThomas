```python
import numpy as np
import math
import matplotlib.pyplot as plt


#print all the numbers that you make a 90 degree turn at
Turningpoint = []
for i in range(100):
    final = math.floor(1/4 * (i + 2)**2 +1)
    Turningpoint.append(final)
    i+=1
print(Turningpoint)


Turns = []
for i in range(2,100):
    final = math.ceil((4 * (i - 1))**0.5 -2)
    Turns.append(final)
    i+=1
print(Turns)



x = np.arange(2,50)
y = np.ceil((4 * (x - 1))**0.5 -2)
plt.plot(x,y,'x')
plt.xlabel('n (Number in spiral)')
plt.ylabel('Number of 90 degree turns')
plt.title('The number of 90 degree turns required to get n')

plt.show()
```

```python
def spiral_coordinates(max_num):
    if max_num < 1:
        return {}

    # Initialize variables
    x, y = 0, 0
    dx, dy = 0, -1
    
    def  right():
        nonlocal dx, dy
        dx, dy = -dy, dx

    coordinates = {}
    for num in range(1, max_num + 1):
        coordinates[num] = (x, y)

        if x == y or (x < 0 and x == -y) or (x > 0 and x == 1 - y):
            right()
            
        x, y = x + dx, y + dy

    return coordinates

max_num = 20
coordinates = spiral_coordinates(max_num)
for num, coord in coordinates.items():
    print(f"Number {num}: Coordinates {coord}")
```
