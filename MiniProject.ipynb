```python
import matplotlib.pyplot as plt

# Define the range for x
x_values = range(1, 101)

# Calculate the corresponding y values
y_values = [sum(1 / (2 * r - 1) for r in range(1, x + 1)) for x in x_values]

# Plot the graph
plt.plot(x_values, y_values)
plt.xlabel('Number of strings in bowl')
plt.ylabel('Expected number of loops formed')
plt.title('Expected number of loops formed for a given N')
plt.grid(True)
plt.show()

# Print the coordinates
print("Coordinates:")
for x, y in zip(x_values, y_values):
    print(f"N = {x}, expected = {y}")
```

```python
import matplotlib.pyplot as plt
from fractions import Fraction

# Define the range for N
x_values = list(range(1, 11))

# Calculate the cumulative product of 1/(2N-1) up to each value of N
y_values = [Fraction(1)]
for x in x_values:
    y_values.append(y_values[-1] * Fraction(1, 2 * x - 1))

# Print the values of y
print("Values of y:")
for x, y in zip(x_values, y_values[1:]):  # Exclude the initial 1
    print(f"x = {x}, y = {y}")

# Plot the graph
plt.plot(x_values, y_values[1:])  # Exclude the initial 1
plt.xlabel('Number of strings')
plt.ylabel('probability all loops formed are size 1')
plt.title('probability all loops formed are size 1 for a given N')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
from fractions import Fraction

# Define the range for n
n_values = list(range(1, 50))

# Calculate the cumulative product of 2n-2 / (2n - 1) up to each value of n
y_values = [1]
for n in n_values:
    product = Fraction(2 * n - 2, 2 * n - 1)
    y_values.append(y_values[-1] * product)
    print(f"n = {n}, Product = {y_values[-1]}")

# Plot the graph
plt.plot(n_values, y_values[1:])  # Exclude the initial 1
plt.xlabel('N')
plt.ylabel('Proabbility of forming a single loop')
plt.title('Probability of forming a single loop for a given N')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Define the range for N
N_values = list(range(1, 101))

# Calculate the sum of 1/r from 1 to N for each value of N
sum_values = [0]
for N in N_values:
    summation = sum(1/r for r in range(1, N + 1))
    sum_values.append(summation)
    print(f"N = {N}, Expected = {summation}")

# Plot the graph
plt.plot(N_values, sum_values[1:])  # Exclude the initial 0
plt.xlabel('N')
plt.ylabel('Sum from 1 to N of 1/r')
plt.title('Plot of Sum from 1 to N of 1/r against N')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Define the range for N
N_values = list(range(1, 101))

# Calculate the sums for each expression
sum_values_1 = [sum(3 / (2*r + 1) for r in range(1, N + 1)) for N in N_values]
sum_values_2 = [sum(2 / (r + 1) for r in range(1, N + 1)) for N in N_values]
sum_values_3 = [sum(1 / (2*r + 1) for r in range(1, N + 1)) for N in N_values]

# Plot the graph
plt.plot(N_values, sum_values_1, label='3 times')
plt.plot(N_values, sum_values_2, label='4 times')
plt.plot(N_values, sum_values_3, label='0.5 times')

plt.xlabel('Number of strings in basket')
plt.ylabel('Expected number of loops formed')
plt.title('Expected number of loops formed for a given N')
plt.legend()
plt.grid(True)
plt.show()
```
